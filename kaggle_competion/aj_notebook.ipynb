```python
import pandas as pd
import numpy as np
from autogluon.tabular import TabularPredictor

directory = './dataset/'  # directory where you have downloaded the data CSV files from the competition
label = 'DiagPeriodL90D'  # name of target variable to predict in this competition
eval_metric = 'roc_auc'  # Optional: specify that competition evaluation metric is AUC
save_path = directory + 'AutoGluonModels/'  # where to store trained models

train_data = pd.read_csv(directory+'training.csv')
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
print (train_data)
```

```python
predictor = TabularPredictor(label=label, eval_metric=eval_metric, path=save_path, verbosity=3).fit(
    train_data, presets='best_quality', time_limit=3600
)

results = predictor.fit_summary()
```

```python
predictor.leaderboard()
```

```python
test_data = pd.read_csv(directory+'test.csv')
y_predproba = predictor.predict_proba(test_data, as_multiclass=False)
y_predproba
```

```python
submission = pd.read_csv('sample_submission.csv')
submission
submission['DiagPeriodL90D'] = y_predproba
submission.head()
submission.to_csv(directory+'my_submission_2.csv', index=False)
```
